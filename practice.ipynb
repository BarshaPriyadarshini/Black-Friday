```python
df.info()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df
```

```python
#handling categorical feature gender
pd.get_dummies(df['Gender'],drop_first=1) # to drop gender
```

```python
df['Gender']=df['Gender'].map({'F':0,'M':1}) # other way
df.head()
```

```python
# handle categorical features age

df['Age'].unique()   # give the unique value
```

```python
pd.get_dummies(df['Age'],drop_first=True)
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7}) 
```

```python
df.head()
```

```python
# fixing categorical city_category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
#drop city category feature
df.drop('City_Category',axis=1,inplace=True)    #axis =1 drop columne wise
```
